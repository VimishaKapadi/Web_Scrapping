```python
Product_name = []
Price = []
Description = []
Review = []
```

```python
for i in range(2,12):
  url = 'https://www.flipkart.com/search?q=washing%20machine%20automatic%20below%2050000&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page='+str(i)
  r = requests.get(url)

  soup = BeautifulSoup(r.text, 'lxml')
  box = soup.find('div',class_ = '_1YokD2 _3Mn1Gg')
  names = box.find_all('div',class_ = '_4rR01T')

  for i in names:
    name = i.text
    Product_name.append(name)

  prices = box.find_all('div',class_ = '_30jeq3 _1_WHN1')

  for i in prices:
    prices = i.text
    Price.append(prices)

  Desc = box.find_all('ul',class_ = '_1xgFaf')

  for i in Desc:
    name= i.text
    Description.append(name)

  review = box.find_all('div',class_ = '_3LWZlK')

  for i in review:
    name = i.text
    Review.append(name)

```

```python
a = {"Product Name":Product_name,"Prices":Price,"Description":Description,"Review":Review}
df = pd.DataFrame.from_dict(a, orient='index')
df = df.transpose()
```
